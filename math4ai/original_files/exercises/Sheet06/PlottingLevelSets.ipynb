```python
import matplotlib.pyplot as plt 
import numpy as np 
  
delta = 0.002    
    
x = np.arange(-1.9, 1.9, delta) 
y = np.arange(-1.2, 1.2, delta) 
[X, Y] = np.meshgrid(x, y) 
  
fig, ax = plt.subplots() 
  
# Z = X**2+2*X*Y+3*Y**2
#This function is given by a quadratic form: it is (x,y)[[1, 1], [1, 3]](x,y)^T
#Diagonalizing this quadratic form as in the principal axis theorem gives, 
#after some rounding, the function
#Z = 3.14*X**2+0.6*Y**2 
#You can check it out that if you plot the contour of this you get a rotated version
#of the above! This may provide some additional intuition about the principal axis theorem.


Z = (X-1)**2-Y**2



#The parameter [1] in the following says that we plot the level set of the value 1
#One could also insert a list of values here. 
#In this caseYou should order them by size, else you get an error.
ax.contour(X, Y, Z, [-2,-1,0,1,2,3,4,5,6])   
ax.set_xlabel('x') 
ax.set_ylabel('y') 

#contours = ax.contour(XPoints, YPoints, ZPoints)
#ax.clabel(contours, inline=1, fontsize=10)
  
plt.show() 
```

```python
import numpy as np
import matplotlib.pyplot as plot
import pylab

# List of points in x axis
XPoints     = []

# List of points in y axis
YPoints     = []

# X and Y points are from -6 to +6 varying in steps of 2 
for val in range(-6, 8, 2):
    XPoints.append(val)
    YPoints.append(val)

# Z values as a matrix
ZPoints     = np.ndarray((7,7))

# Populate Z Values (a 7x7 matrix) - For a circle x^2+y^2=z    
for x in range(0, len(XPoints)):
    for y in range(0, len(YPoints)):
        ZPoints[x][y] = (XPoints[x]* XPoints[x]) + (YPoints[y]*YPoints[y])

# Print x,y and z values
print(XPoints)
print(YPoints)
print(ZPoints)

# Set the x axis and y axis limits
pylab.xlim([-10,10])
pylab.ylim([-10,10])

# Provide a title for the contour plot
plot.title('Contour plot')

# Set x axis label for the contour plot
plot.xlabel('X')

# Set y axis label for the contour plot
plot.ylabel('Y')

# Create contour lines or level curves using matplotlib.pyplot module
contours = plot.contour(XPoints, YPoints, ZPoints)

# Display z values on contour lines
plot.clabel(contours, inline=1, fontsize=10)

# Display the contour plot
plot.show()
```
