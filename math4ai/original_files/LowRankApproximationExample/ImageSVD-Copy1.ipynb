```python
#from: https://www.frankcleary.com/svdimage/
import matplotlib.pyplot as plt
import numpy as np
import time
from PIL import Image
```

```python
img = Image.open('Cat.jpg')
imggray = img.convert('LA')
plt.figure(figsize=(9, 6))
plt.imshow(imggray);
```

```python
#Convert the image data into a numpy matrix, plotting the result to show the data is unchanged.
imgmat = np.array(list(imggray.getdata(band=0)), float)
imgmat.shape = (imggray.size[1], imggray.size[0])
imgmat = np.matrix(imgmat)
plt.figure(figsize=(9,6))
plt.imshow(imgmat, cmap='gray');
```

```python
#Now to compute the singular value decomposition:
U, sigma, V = np.linalg.svd(imgmat)

reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
plt.imshow(reconstimg, cmap='gray');
```

```python
# for np.matrices, the operator * is matrix multiplication
A = np.random.randint(0, 100, (3,3))
B = np.random.randint(0, 100, (3,3))
C = np.matrix(A)
D = np.matrix(B)
print(A==C)
print(B==D)
```
