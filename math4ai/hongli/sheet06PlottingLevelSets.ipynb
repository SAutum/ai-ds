```python
import matplotlib.pyplot as plt
import numpy as np

step = 2e-2
x = np.arange(-9,9, step)
y = np.arange(-1,30, step)
[X, Y] = np.meshgrid(x, y)
fig, ax = plt.subplots()
F = (X-3)**2 + (Y-1)**2
ax.axis('equal')
cp = ax.contour(X, Y, F, colors ="black", linestyles = "dashed", alpha = 0.5)
ax.clabel(cp, inline=1, fontsize=10)
ax.contourf(X, Y, F, alpha = 0.5)
ax.scatter([3], [1], c = "black")

ax.plot(x, x**2 +1, c= "black")
ax.plot(x, -3*x + 11)
ax.grid()
ax.set_xlim(-3, 9)
ax.set_ylim(-1, 30)
plt.show()
```

```python
step = 2e-2
x = np.arange(-15,15, step)
y = np.arange(-5,20, step)
[X, Y] = np.meshgrid(x, y)
fig, ax = plt.subplots()
F = -4*X*Y + 3*X**2 + 2*X + 4*Y
ax.axis('equal')
cp = ax.contour(X, Y, F, colors ="black", linestyles = "dashed", alpha = 0.5)
ax.clabel(cp, inline=1, fontsize=10)
ax.contourf(X, Y, F, alpha = 0.5)

ax.plot(np.ones_like(y)*(-1), y, c= "black")
ax.plot(np.ones_like(y)*(3), y, c= "black")
ax.plot(x,np.zeros_like(x), c= "black")

ax.grid()
ax.set_xlim(-15, 15)
ax.set_ylim(-5, 20)
plt.show()
```
