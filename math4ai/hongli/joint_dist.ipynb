```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
X = norm(loc = 1)
Y = norm(loc = 1.5, scale = 2)
```

```python
fig, ax = plt.subplots()
x = np.linspace(X.ppf(0.01), X.ppf(0.99), 100)
plt.plot(x, X.pdf(x), label = "PDF - X")
plt.plot(x, Y.pdf(x), label = "PDF - Y")

plt.legend()
plt.show()
```

```python
[X_density, Y_density] = np.meshgrid(X.pdf(x), Y.pdf(x))
fig, ax = plt.subplots()
F = X_density*Y_density
ax.plot(x, -3/4*x + 2, label = "z = 3*x + 4*y = 8")
print("Probability of the whole area: 1")
ax.contourf(x, x, F, alpha = 0.5)
ax.legend()
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid()
fig.show()
```
