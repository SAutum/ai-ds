```python
# Open and read the content of the text
# The content will remain after closing the file
path = "data/hhu_wikipedia.txt"
file = open(path)
content = file.read()
file.close()
```

```python
content
```

```python
# There are lots of unwanted characters to be cleaned.
# The capitalized letters can also be lowered.
import re

cleaned_content = re.sub(r"\'s", " ", content) # HHU's/Düsseldorf's
cleaned_content = re.sub(r"[^a-zA-ZäöüÄÖÜ\s]", " ", content)

# cleaned_content = cleaned_content.lower()

# Split the cleaned data to a list of words
cleaned_content = cleaned_content.split()
cleaned_content
```

```python
cleaned_content.count("Mathematics")
```

```python
# Extract the unique words as keys for the "count" method
keys = list(set(cleaned_content))
frequencies = [cleaned_content.count(key) for key in keys]

for x in range(10):
    # Find the index of the word with highest frequency
    index = frequencies.index(max(frequencies))
    # Output the exact word through the index
    print("No. {} Word: \"{}\" Frequency: {}".format(x+1, keys[index], frequencies[index]))
    # Remove/Pop the word out of the list
    frequencies.pop(index)
    keys.pop(index)
```

```python
# try to construct function by oneself and compare with numpy built-in functions
# 3x3 only
import numpy as np

A = np.random.random((3,3))

def transpose(A):
    B = A.copy()
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            A[i,j] = B[j, i]
    return B

print(transpose(A))
print(A.T)
```

```python
def determinant_3(A):
    det = 0
    for i in range(3):
        det += A[0,i%3]*A[1,(i+1)%3]*A[2,(i+2)%3]
        det -= A[0,i%3]*A[1,(i-1)%3]*A[2,(i-2)%3]
    return det

print(determinant_3(A))
print(np.linalg.det(A))
```

```python
def matmul(A, B):
    assert A.shape[1] == B.shape[0]
    C = np.empty((A.shape[0],B.shape[1]))
    for i in range(C.shape[0]):
        for j in range(C.shape[1]):
            C[i, j] = sum(A[i,:]*B[:, j])
    return C


print(matmul(A, B))
print(A@B)
```
